### Read tables from silver layer

```python
customers_dim = spark.read.table('`bike-data-lakehouse-workspace`.silver.crm_customer_info')
products_dim = spark.read.table('`bike-data-lakehouse-workspace`.silver.crm_product_info')
fact_product_sales = spark.read.table('`bike-data-lakehouse-workspace`.silver.crm_sales_details')
```

### Replace customer_id with customer_key and drop unwanted columns

```python
fact_product_sales = fact_sales.join(customers_dim.select("Customer_Key", "Customer_id"), "Customer_id", "left")
fact_product_sales = fact_sales.drop("Customer_id", "due_date", "order_dt", "ship_dt", )

display(fact_sales)
```

### Write to gold layer

```python
customers_dim.write.mode('overwrite').saveAsTable('`bike-data-lakehouse-workspace`.gold.DIM_Customer')
products_dim.write.mode('overwrite').saveAsTable('`bike-data-lakehouse-workspace`.gold.DIM_Product')
fact_product_sales.write.mode('overwrite').saveAsTable('`bike-data-lakehouse-workspace`.gold.FACT_Peoduct_Sales')
```
