### read raw data

```python
from pyspark.sql.functions import trim, col
df = spark.read.table('`bike-data-lakehouse-workspace`.bronze.erp_erp_px_cat_g1v2')
```

### rename and trim white spaces

```python
df = df.withColumn('CAT', trim(col('CAT')))
df = df.withColumn('SUBCAT', trim(col('SUBCAT')))
df = df.withColumn('Maintenance', trim(col('Maintenance')))
```

```python
df = df.withColumnRenamed('CAT', 'Category')
df = df.withColumnRenamed('SUBCAT', 'Subcategory')
```

### Write to silver layer

```python
df.write.mode('overwrite').saveAsTable('`bike-data-lakehouse-workspace`.silver.erp_Pricing_Category')
```
