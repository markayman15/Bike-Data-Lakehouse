### read raw data

```python
from pyspark.sql.functions import col, create_map, lit, coalesce
from itertools import chain

df = spark.read.table('`bike-data-lakehouse-workspace`.bronze.erp_locations_a101')
```

### revision CNTRY Column

```python
dic = {
    'US' : 'United States of America',
    'DE' : 'Germany',
}

mapped_dic = create_map(*[lit(x) for x in chain(*dic.items())])

df = df.withColumn('CNTRY', coalesce(mapped_dic.getItem(col('CNTRY')), col('CNTRY')))

print(df.count())
df = df.distinct()
print(df.count())
```

### rename columns

```python
df =df.withColumnRenamed('CNTRY', 'country')
```

### write to silver layer

```python
df.write.mode('overwrite').saveAsTable('`bike-data-lakehouse-workspace`.silver.erp_locations')
```
