```python
from pyspark.sql.functions import when, trim

df = spark.read.table("`bike-data-lakehouse-workspace`.bronze.crm_cust_info")
```

```python
df = df.withColumn('cst_marital_status', when(df.cst_marital_status == 'S', 'Single')
                   .when(df.cst_marital_status == 'M', 'Married'))

df = df.withColumn('cst_gndr', when(df.cst_gndr == 'M', 'Male')
                   .when(df.cst_gndr == 'F', 'Female'))

df = df.withColumn('cst_firstname', trim(df.cst_firstname))
df = df.withColumn('cst_lastname', trim(df.cst_lastname))

df.display()
```

```python
df = (
    df.withColumnRenamed('cst_id', 'Customer_id')
    .withColumnRenamed('cst_key', 'Customer_Key')
    .withColumnRenamed('cst_firstname', 'Customer_Firstname')
    .withColumnRenamed('cst_lastname', 'Customer_Lastname')
    .withColumnRenamed('cst_marital_status', 'Customer_Marital_Status')
    .withColumnRenamed('cst_gndr', 'Customer_Gender')
    .withColumnRenamed('cst_create_date', 'Customer_Create_Date')
)

df.display()
```

```python
df.write.mode('overwrite').saveAsTable('`bike-data-lakehouse-workspace`.silver.crm_customer_info')
```
