### Read Data


```python
from pyspark.sql.functions import col, try_to_date, trim, regexp_replace
df = spark.read.table('`bike-data-lakehouse-workspace`.bronze.crm_sales_details')
```

### Cast Columns to Date type

```python
df = df.withColumn("sls_due_dt", try_to_date(trim(col("sls_due_dt").cast("string")), 'yyyyMMdd'))
df = df.withColumn("sls_ship_dt", try_to_date(trim(col("sls_ship_dt").cast("string")), 'yyyyMMdd'))
df = df.withColumn("sls_order_dt", try_to_date(trim(col("sls_order_dt").cast("string")), 'yyyyMMdd'))
```

### cast columns to double

```python
df = df.withColumn("sls_sales", col("sls_sales").cast("double"))
df = df.withColumn("sls_price", col("sls_price").cast("double"))
```

### rename columns

```python
df = df.withColumnRenamed('sls_prd_key', 'product_key')
df = df.withColumnRenamed('sls_cust_id', 'customer_id')
df = df.withColumnRenamed('sls_order_num', 'order_num')
df = df.withColumnRenamed('sls_ship_num', 'ship_num')
df = df.withColumnRenamed('sls_ship_dt', 'ship_dt')
df = df.withColumnRenamed('sls_due_dt','due_date')
df = df.withColumnRenamed('sls_sales', 'sales')
df = df.withColumnRenamed('sls_price', 'price')
df = df.withColumnRenamed('sls_order_dt', 'order_dt')
df = df.withColumnRenamed('sls_quantity', 'quantity')
```

### write to silver layer

```python
df.write.mode("overwrite").saveAsTable("`bike-data-lakehouse-workspace`.silver.crm_sales_details")
```
