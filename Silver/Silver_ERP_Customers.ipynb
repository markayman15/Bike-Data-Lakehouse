### Read Raw data

```python
from pyspark.sql.functions import when

df = spark.read.table('`bike-data-lakehouse-workspace`.bronze.erp_customers_az12')
```

### Rename columns

```python
df = df.withColumnRenamed('BDATE', 'BIRTHDATE')
df = df.withColumnRenamed('GEN', 'GENDER')
df = df.withColumnRenamed('CID', 'CUSTOMER_ID')
```

### handle wrong values in GENDER column

```python
df = df.withColumn('GENDER', when(df.GENDER == 'M', 'Male')
                   .when(df.GENDER == 'F', 'Female'))
```

### write to silver layer

```python
df.write.mode('overwrite').saveAsTable('`bike-data-lakehouse-workspace`.silver.erp_customers')
```
