# Read and write the raw data from source volume

```python
from source_paths import BRONZE_RAW_SOURCES

for fileName, csvFileMetadata in BRONZE_RAW_SOURCES.items():
    df = spark.read.option('header', 'true').option('inferSchema', 'true').csv(csvFileMetadata['path'])
    df.write.mode('overwrite').saveAsTable(f"`bike-data-lakehouse-workspace`.bronze.{csvFileMetadata['table']}")
```
